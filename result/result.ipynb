```python
import pandas as pd
import matplotlib.pyplot as plt

result_df = pd.read_csv('q-learning-result.csv', index_col='episode')
result_df
```

```python
result_df.describe()
```

```python
scores = result_df['score']
scores_100steps_ma = scores.rolling(window=100).mean()
scores_100steps_ma.dropna(how='any')
```

```python
result_df.plot(kind='line', y='score')
scores_100steps_ma.plot(kind='line', y='score')
plt.legend(['Score', 'Moving Average'])
```
